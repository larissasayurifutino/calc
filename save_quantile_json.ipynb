```python
import pandas as pd
import numpy as np
import json
```

```python
alc = pd.read_csv('data.csv',sep=',', encoding='latin1', decimal=".")
```

```python
alc = alc[['Location', 'Dim1', 'FactValueNumeric']]
alc_fem = alc.query('Dim1 == "Female"')
alc_male = alc.query('Dim1 == "Male"')
```

```python
percentile = np.arange(0, 1, 0.01)
percentile
```

```python
fem_perc = alc_fem['FactValueNumeric'].quantile(percentile)
fem_perc = pd.DataFrame(fem_perc)
fem_perc
```

```python
alc_fem.shape[0]
#80/30
#27/185
```

```python
alc_fem[alc_fem['FactValueNumeric'] < 2.6]
```

```python
arr_fem_perc = []
for index, row in fem_perc.iterrows():
    obj = {
    'quantil': int(row.name * 100),
    'valor': round(row.FactValueNumeric, 3)
    }

    arr_fem_perc.append(obj) 
```

```python
arr_fem_perc
```

```python
male_perc = alc_male['FactValueNumeric'].quantile(percentile)
male_perc = pd.DataFrame(male_perc)
male_perc
```

```python
arr_male_perc = []
for index, row in male_perc.iterrows():
    obj = {
    'quantil': int(row.name * 100),
    'valor': round(row.FactValueNumeric, 3)
    }

    arr_male_perc.append(obj)
```

```python
jsonString = json.dumps(arr_fem_perc)
jsonFile = open("fem_perc.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```

```python
jsonString = json.dumps(arr_male_perc)
jsonFile = open("male_perc.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```
